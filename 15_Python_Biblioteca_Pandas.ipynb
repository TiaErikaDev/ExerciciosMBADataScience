# **Aula 16: Seleção de Dados em Pandas**


```python
import pandas as pd
import numpy as np
!pip install names
import names
```

```python
n = 200
nomes = [names.get_full_name() for i in range(n)]
p1 = np.random.uniform(0, 7, n)
p2 = np.random.uniform(3, 10, n)
p3 = np.random.uniform(1, 9, n)

dicionario = {'Nomes': nomes,
              'p1': p1,
              'p2': p2,
              'p3': p3}

notas = pd.DataFrame(dicionario).round(1)

notas['MF'] = np.ceil(10*notas.mean(axis = 1))/10

def auxiliar(nota):
    if nota >= 6:
      saida = 'Aprovado'
    else:
      saida = 'Reprovado'
    return saida

notas['Situação'] = notas['MF'].apply(auxiliar)
notas                                      
```
