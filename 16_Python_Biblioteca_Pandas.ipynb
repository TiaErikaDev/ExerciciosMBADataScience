# **Aula 17 - Agrupamento no Pandas**

```python
import pandas as pd
import numpy as np
!pip install names
import names
```

```python
n = 200
nomesM = [names.get_full_name(gender = 'male') for i in range(int(n/2))]
nomesF = [names.get_full_name(gender = 'female') for i in range(int(n/2))]
freq = (np.random.uniform(0, 1, n)*100).round()

nomes = nomesM + nomesF
sexo = ['M']*int(n/2) + ['F']*int(n/2)
p1 = np.random.uniform(0, 7, n)
p2 = np.random.uniform(3, 10, n)
p3 = np.random.uniform(1, 9, n)

dicionario = {'Nomes': nomes,
              'Sexo': sexo,
              'Frequência': freq,
              'p1': p1,
              'p2': p2,
              'p3': p3}

notas = pd.DataFrame(dicionario).round(1)

notas['MF'] = ((p1 + p2 + p3)/3).round(1)

def aux(notas):
  if notas > 40.0:
    saida = 'Média Frequência'
    if notas > 70.0:
      saida = 'Alta Frequência'
  else:
    saida = 'Baixa Frequência'
  return saida
notas['Sit'] = notas['Frequência'].apply(aux)

def auxiliar(nota):
    if nota >= 6:
      saida = 'Aprovado'
    else:
      saida = 'Reprovado'
    return saida
notas['Situação'] = notas['MF'].apply(auxiliar)

notas 
```

```python
notas[notas['Sexo'] == 'M']['MF'].mean()
```

```python
notas[notas['Sexo'] == 'F']['MF'].mean()
```

```python
notas['MF'].groupby(notas['Sexo']).mean()
```

```python
notas.groupby(['Situação', 'Sexo']).mean()
```
