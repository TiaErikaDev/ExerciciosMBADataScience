### **Aula 20 - Webscrapping com Pandas**

```python
import pandas as pd
```

```python
salvador = pd.read_html('https://pt.wikipedia.org/wiki/Salvador', thousands = '.')
```

```python
#Tipo da tabela gerada
type(salvador)
```

```python
#números de tabelas
len(salvador)
```

```python
#Tabela de interesse
salvador[4]
```

# Limpeza dos dados da tabela

```python
#Criando o objeto com a tabela de interesse
dados = salvador[4]
dados
```

```python
#Organizando as colunas com os meses do ano
dados.columns = [dados.columns[i][1] for i in range(len(dados.columns))]
dados
```

```python
#Colocando a primeira coluna como index(índice)
dados.index = dados['Mês']
```

```python
#Excluir a coluna Mês
dados.drop('Mês', axis = 1, inplace = True)
dados
```

```python
#Excluir a última linha desnecessária
dados = dados.iloc[:-1, :]
dados
```

```python
#Mudar o tipo de dado com ',' para '.'
for coluna in dados.columns:
  dados[coluna] = dados[coluna].str.replace(',', '.')
```

```python
#tirando o espeço entre números ('\xa0')
dados['Ano'] = dados['Ano'].str.replace('\xa0', '')
```

```python
#Mudar o tipo de dado str para float
for coluna in dados.columns:
  dados[coluna] = dados[coluna].astype(float)
```

```python
dados.describe()
```

```python
#criando gráficos de linha
dados.iloc[:5, :12].T.plot(ylim = (0, 38));
```
